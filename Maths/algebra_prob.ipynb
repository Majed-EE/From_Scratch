```python
######## chapter 16 #### statistics mit
sides=[4,6,8,12,20]
no_of_h=len(sides)
H = list([1/no_of_h for x in range(no_of_h) ]) # prior probability assuming flat prior
D=[4,6,8,12,20]

def posterior(D,H): # posterior with data and hypothesis
    prior_H=H
    post_h=[]
    for step in range(len(D)):
        posterior_H=[]
        data=D[step]
        likelihoods = list(prior_H[x]*1/sides[x] if sides[x] >= data else 0 for x in range(len(H)))
        P_D = sum(likelihoods[x] for x in range(len(likelihoods)))
        for x in range(len(H)):
            prior=prior_H[x] # priod
            norm=P_D # probability of data
            post_p=likelihoods[x]*prior/norm
#             print(f"data")
            posterior_H.append(post_p)
        prior_H=posterior_H
        post_h.append(posterior_H)
    return post_h
        
```

```python
# Generate the data
import numpy as np
from matplotlib import pyplot as plt
n=2000


x=np.random.uniform(-np.pi,np.pi,n*2).reshape(n,2)
# decision boundary f(x)=pi sin(x)
```

```python
def f(x):
    return np. pi*np.sin(x)
y=(x[:,1]>f(x[:,0])).reshape(-1,1)
line_x=np.arange(-np.pi,np.pi,0.1)
line_y=f(line_x)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x[:,0],x[:,1],c=y,s=7)
plt.plot(line_x, line_y,c='red')
```

```python
# create nn of fixed size

class NN():
    def __init__(self):
        self.b1=np.random.randn(1,3)
        self.W1=np.random.randn(2,3)
        self.b2=np.random.randn(1,1)
        self.W2=np.random.randn(3,1)
    def sigmoid(self,z):
        return np.exp(z)/(1+np.exp(z))
    
    def predict(self,x):
        # forward prop
        z1=x@self.W1+self.b1
        a1=self.sigmoid(z1)
        z2=a1@self.W2+self.b2
        y_hat=self.sigmoid(z2)
        return y_hat
    
    def loss(self,y,y_hat):
        loss=-np.sum(y*np.log(y_hat)+(1-y)*np.log(1-y_hat)) #BCE
        return loss
    def train(self,x,y,epochs=50,learning_rate=0.001):
        losses=[]
        for j in range (epochs):
            # forward prop, n is the batch size
            z1=x@self.W1+self.b1  # (n,3)
            a1=self.sigmoid(z1)   # (n,3)
            z2=a1@self.W2+self.b2 # (n,1)
            a2=self.sigmoid(z2)   # (n,1)
            
            # keep track of the loss
            loss=self.loss(y,a2) # 
            losses.append(loss)
            
            
            # backprop
            dL_dz2=(a2-y)                  # (n,1)
            dL_dW2=a1.T@dL_dz2             # (3,n)x(n,1)=(n,3)
            ones=np.ones((n,1)) # (1,n)x(n,1)=(1,1)
            dL_db2=ones.T@dL_dz2      
            print(dL_dz2.shape, (self.W2.T).shape)
            delta1=dL_dz2@(self.W2.T)       # ???(nx1)x(1,3)=(n,3)
            sig_der=self.sigmoid(a1)*(1-self.sigmoid(a1))
            dL_dW1=x.T@(delta1*sig_der)    # (2,n)x(n,3)=(2,3)
            dL_db1=ones.T@(delta1*sig_der)  # (1,n)x(n,3)=(1,3)
            
            #SGD
            self.b1=self.b1-learning_rate*dL_db1
            self.W1=self.W1-learning_rate*dL_dW1
            self.b2=self.b2-learning_rate*dL_db2
            self.W2=self.W2-learning_rate*dL_dW2
            
        return losses
            
            
    
    
    
```

```python
nn=NN()
y_hat_bt=nn.predict(x)
```

```python
pred=y_hat_bt>0.5
acc=(y==pred).mean()
print(f"accuracy = ={acc}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Scenario 1: High variance along x-axis
x_high_var = np.random.normal(loc=0, scale=5, size=100)  # High variance (scale=5)
y_high_var = np.random.normal(loc=0, scale=1, size=100)  # Low variance in y

# Scenario 2: Low variance along x-axis
x_low_var = np.random.normal(loc=0, scale=0.5, size=100)  # Low variance (scale=0.5)
y_low_var = np.random.normal(loc=0, scale=1, size=100)     # Same y-variance as Scenario 1

# Create subplots for comparison
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot Scenario 1 (High x-variance)
ax1.scatter(x_high_var, y_high_var, color='red', alpha=0.7)
ax1.set_title("High Variance Along X-Axis", fontsize=14)
ax1.set_xlabel("X (σ² = 25)", fontsize=12)
ax1.set_ylabel("Y (σ² = 1)", fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.6)
ax1.set_xlim(-15, 15)
ax1.set_ylim(-4, 4)

# Plot Scenario 2 (Low x-variance)
ax2.scatter(x_low_var, y_low_var, color='blue', alpha=0.7)
ax2.set_title("Low Variance Along X-Axis", fontsize=14)
ax2.set_xlabel("X (σ² = 0.25)", fontsize=12)
ax2.set_ylabel("Y (σ² = 1)", fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.6)
ax2.set_xlim(-15, 15)
ax2.set_ylim(-4, 4)

plt.tight_layout()
plt.show()
```
