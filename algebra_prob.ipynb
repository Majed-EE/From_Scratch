```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Define parameters m and n
m = 2
n = 2


# Generate x values
x = np.linspace(0, 1, 100)

# Compute the beta PDF
y = beta.pdf(x, m, n)

# Plot the beta distribution
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'Beta Distribution (m={m}, n={n})', color='blue')
plt.title('Beta Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Define parameters m and n
m = 2
n = 2


# Generate x values
x = np.linspace(0, 1, 100)
y2=x
# Compute the beta PDF
y = beta.pdf(x, m, n)*x

# Plot the beta distribution
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'Beta Distribution (m={m}, n={n})', color='blue')
plt.title('Beta Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
n=np.linspace(0,1,100)
prob=np.zeros((n.size))
for n_current in range (n.size):
    pdu=1
    for x_current in range (n.size):
         pdu=pdu*(n[n_current]**x[x_current])*((1-n[n_current])**(1-x[x_current]))
      
    prob[n_current]=pdu
    print(pdu)
    
    
    
```

```python
# Plot the beta distribution
plt.figure(figsize=(8, 6))
plt.plot(n, prob, label=f'Beta Distribution (m={m})', color='blue')
plt.title('Beta Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
