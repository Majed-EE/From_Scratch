```python
import numpy as np


mean_X = 3
var_X = 4
std_X = np.sqrt(var_X)
X = np.random.normal(loc=mean_X, scale=std_X, size=1000)  # 1000 samples

mean_Y = 8
var_Y = 2
std_Y = np.sqrt(var_Y)
Y = np.random.normal(loc=mean_Y, scale=std_Y, size=1000)

# Random variable that is the sum of the above two random variables
Z = X + Y

# Calculate properties of Z
mean_Z = np.mean(Z)
# mean is the direct sum 
# U_z=U_x+U_y


var_Z = np.var(Z)

print(f"Generated random variable X: mean={np.mean(X):.2f}, variance={np.var(X):.2f}")
print(f"Generated random variable Y: mean={np.mean(Y):.2f}, variance={np.var(Y):.2f}")
print(f"Sum random variable Z: mean={mean_Z:.2f}, variance={var_Z:.2f}")
```

```python
######## chapter 16 #### statistics mit
sides=[4,6,8,12,20]
no_of_h=len(sides)
H = list([1/no_of_h for x in range(no_of_h) ]) # prior probability assuming flat prior
D=[4,6,8,12,20]

def posterior(D,H): # posterior with data and hypothesis
    prior_H=H
    post_h=[]
    for step in range(len(D)):
        posterior_H=[]
        data=D[step]
        likelihoods = list(prior_H[x]*1/sides[x] if sides[x] >= data else 0 for x in range(len(H)))
        P_D = sum(likelihoods[x] for x in range(len(likelihoods)))
        for x in range(len(H)):
            prior=prior_H[x] # priod
            norm=P_D # probability of data
            post_p=likelihoods[x]*prior/norm
#             print(f"data")
            posterior_H.append(post_p)
        prior_H=posterior_H
        post_h.append(posterior_H)
    return post_h
        
```

```python
k=posterior(D,H)
```

```python
import scipy.stats as stats

# Define the parameters of the beta distribution
a, b = 11, 12

# Calculate the probabilities for the intervals
prob_0_3_to_0_65 = stats.beta.cdf(0.65, a, b) - stats.beta.cdf(0.3, a, b)
prob_0_42_to_0_57 = stats.beta.cdf(0.57, a, b) - stats.beta.cdf(0.42, a, b)

print(f"P([0.3, 0.65]) = {prob_0_3_to_0_65:.2f}")
print(f"P([0.42, 0.57]) = {prob_0_42_to_0_57:.2f}")
```

```python
import numpy as np
from matplotlib import pyplot as plt
x = np.linspace(0, 1, 1000)
pdf_values = stats.beta.pdf(x, a, b)

# Plot the beta distribution
plt.plot(x, pdf_values, label=f'Beta({a}, {b})')

# Highlight the intervals
plt.axvline(x=0.3, color='blue', linestyle='--', label='0.3')
plt.axvline(x=0.65, color='blue', linestyle='--', label='0.65')
plt.axvline(x=0.42, color='orange', linestyle='--', label='0.42')
plt.axvline(x=0.57, color='orange', linestyle='--', label='0.57')
plt.fill_between(x, pdf_values, where=(x >= 0.3) & (x <= 0.65), color='blue', alpha=0.3)
plt.fill_between(x, pdf_values, where=(x >= 0.42) & (x <= 0.57), color='orange', alpha=0.3)

plt.title('Beta Distribution with Highlighted Intervals')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()

```
